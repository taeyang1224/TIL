# 데이터, 데이터베이스, DBMS

## 데이터란?
- 컴퓨터가 처리할 수 있는 문자, 숫자, 소리, 그림 따위의 형태로 된 정보

## 데이터베이스란?
- 데이터 저장소 (여러 사람이 공유하고 사용할 목적으로 통합 관리 저장소)
    - 테이블을 저장해두는 저장소 == 스키마
- 자료를 구조화하여 저장하기 때문에 효율적인 관리가 가능
- 여러 업무에 여러 사용자가 동시에 사용 가능
- 데이터베이스의 기능을 사용하기 위해 응용 프로그램을 활용

## DBMS란?
- Database Management System
- 데이터베이스의 기능을 제공하는 프로그램

### DBMS의 종류
- Oracle, MySQL, MSQL, MariaDB

# 데이터베이스 언어와 SQL

## SQL
- 데이터베이스와 대화하기 위해 사용하는 언어

## 쿼리
- SQL로 쓰인 데이터베이스에 명령을 내리는 문장

# 데이터 종류 (숫자, 문자, 그림, 영상 data)
- 데이터를 저장하기 전에 저장 공간의 데이터 타입을 미리 정해줘야함
- 해당 저장 공간에서 미리 정해둔 데이터 타입이 아닐경우, *ERROR* 발생

## 숫자형 데이터 타입
- 정수형: 소수점이 없는 데이터 

- 실수형: 소수점이 있는 데이터 

## 문자형: 텍스트로 구성된 문자열 데이터
- Char의 경우 고정길이 데이터, varchar의 경우 변동길이 데이터


## 날짜형: 날짜와 시간 데이터


# 기본적인 데이터 다루기

## 숫자형 데이터 기본 특징
- 숫자형 데이터는 "수", 데이터 간 연산이 가능

## 문자형 데이터 기본 특징
- 문자형 데이터는 반드시 " or '로 사용
- 따옴표가 없는 문자는 키워드 or 함수, 데이터베이스 / 테이블 / 컬럼의 이름으로 인식

## 데이터 타입 간 타입 변환
- 숫자형, 문자형, 날짜형 데이터는 함수를 사용하여 서로 타입 변환이 가능

# 테이블 만들고 변경하기

## 테이블이란?
- 데이터베이스에서 데이터의 형태를 정해 모아 놓은 저장 공간 **(행과 열로 이루어진 데이터 표)**

### 컬럼 (열)
- 데이터를 저장하기 위한 틀
    - 컬럼의 이름과 데이터 타입은 테이블을 만들 때 **미리 정함**
    - 컬럼의 이름은 동일한 테이블 내에서 **중복 x**
    - 테이블은 **반드시 1개 이상의 컬럼**을 갖고 있어야 함

### 값
- 컬럼에 속한 실제 데이터 값
    - **컬럼의 데이터 타입** 만을 값으로 가질 수 있음

### 로우 (행)
- 관계된 값의 리스트
    - 하나의 로우는 하나의 관계된 데이터를 의미 **(하나의 로우가 한 사람의 데이터)**
    - 같은 테이블 안에서 로우는 **항상 동일한 구조**를 가짐
    - **로우를 단위로 데이터를 삽입**
    
## 테이블과 데이터베이스의 생성 순서
- 데이터베이스를 생성한 후 특정 뎅터베이스 내에 테이블을 생성


## 데이터베이스, 테이블, 컬럼 이름 정하기 규칙
1. 문자, 숫자, _ 사용
2. 이름에 쓰이는 문자는 주로 영문 소문자를 사용
    - 한글도 사용하지만 `인코딩 이슈`로 주로 영문 사용
    - 보통 키워드나 함수명은 대문자, 사용자가 정의한 이름에는 소문자 사용
3. 예약어는 사용 X
    예약어: 이미 키워드, 함수명 등의 문법적인 용도로 사용되고 있기 때문에 이름으로 사용 X
4. 단어와 단어 사이에는 빈칸 대신 _ 를 사용
5. 문자로 시작합니다. (숫자, _ 로 시작 X)
6. 데이터베이스 이름은 중복 X
    - 테이블 이름은 하나의 데이터베이스 내에서는 중복 X
    - 컬럼 이름은 하나의 테이블 내에서는 중복 X

# 문법
- CREATE DATABASE : 데이터베이스를 만듬
- CREATE TABEL : 테이블을 만듬
    - USE를 사용 안할시 **CREATE TABEL 데이터베이스.테이블**
- INSERT INTO 테이블명 (컬럼1, 컬럼2 ...) VALUES (로우1, 로우2 ...) : 테이블에 값을 추가
- ALTER TABEL 테이블명 RENAME 테이블명 : 테이블명 변경
- ALTER TABLE 테이블명 CHANGE COLUMN 컬럼명 타입 : 컬럼명 변경 **(이름만 변경은 안됨, 타입도 같이 변경해야함)**
- TRUNCATE 테이블명 : 테이블은 그대로 두고 value만 다 지움
- DROP TABLE 테이블명: 테이블을 삭제
