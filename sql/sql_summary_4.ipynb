# 원하는 데이터 만들기

## ORDER BY
- 가져온 데이터를 정렬해주는 키워드
- `ORDER BY [컬럼]` 형식으로 사용
- 입력한 컬럼 이름의 값ㅇ르 기준으로 모든 ROW를 정렬
- 기본 정렬 규칙은 **오름차순** `ORDER BY [컬럼] = ORDER BY [컬럼] ASC` 내림차순은 `DESC` 사용
- 여러 컬럼 정렬 가능
- `ORDER BY [컬럼1], [컬럼2]...` 왼쪽부터 차례대로 정렬
- 컬럼 번호로 정렬 가능

## RANK
- 데이터를 정렬해 순위를 만들어주는 함수
- `RANK() OVER (ORDER BY [컬럼])` 형식으로 사용
- 항상 `ORDER BY`와 함께 사용
- SELECT절에 사용, 정렬된 순서에 순위를 붙인 새로운 컬럼을 보여줌 **(테이블의 실제 데이터에는 영향 X)**
### 같은 RANK 다른 OUTPUT
- RANK: 공동 순위가 있으면 다음 순서로 건너 뜀
- DENSE_RANK: 공동 순위가 있어도 다음 순위를 뛰어 넘지 않음
- ROW_NUMBER: 공동 순위를 무시함


## 함수
- `함수 이름(함수를 적용할 값 OR 컬럼 이름)` 형식으로 사용
- 결과 값을 새로운 컬럼으로 반환

### 자주사용하는 함수

#### 문자형 데이터
- LOCATE
    - `LOCATE("A", "B")` B에서 A는 몇번째에 위치해 있는지 검색하여 위치 반환
    - 문자가 여러개라면 가장 먼저 찾은 **문자의 위치**를 가져옴 **(만약 찾는 문자가 없다면 0을 가져옴)**
- SUBSTRING
    - `SUBSTRING("A", 2)` A에서 2번째 문자부터 반환
    - 입력한 숫자가 문자열의 길이보다 크다면 아무것도 가져오지 않음
- LEFT, RIGHT
    - `LEFT("A", 1) or RIGHT("A", 1)` A에서 왼쪽 OR 오른쪽에서 1번째 문자까지 반환
- UPPER, LOWER
    - `UPPER("A") or LOWER("A")` A를 대문자 or 소문자로 바꿔 반환
- LENGTH
    - `LENGTH("A")` A의 글자 수를 변환
- CONCAT
    - `CONCAT("A", "B")` A 문자열과 B 문자열을 합쳐 반환

#### 숫자형 데이터
- ABS
    - `ABS(숫자)` 숫자의 절댓값 반환
    - FLOAT 데이터 타입은 입력 값의 근사치를 저장하기 때문에 근사값으로 반환
- CEILING, FLOOR
    - `CEILING(숫자) or FLOOR(숫자)` 숫자를 정수로 올림 or 내림해서 반환
- ROUND, TRUNCATE
    - `ROUND(숫자, 자리수) OR TRUNCATE(숫자, 자리수)` 숫자를 소수점 자릿수까지 반올림 or 버림해서 반환
    - 자릿수에 0 입력 시 소수점 없이 정수만 반환
- POWER
    - `POWER(A, B)` A의 B 제곱 반환
- MOD
    - `MOD(A, B)` A를 B로 나눈 나머지 반환

#### 날짜형 데이터
- NOW, CURRENT_DATE, CURRENT_TIME
    - 입력값이 필요 없음
    - 현재 날짜와 시간 반환
- YEAR, MONTH, MONTHNAME
    - 날짜의 연도, 월 반환
    - MONTHNAME은 영어로 반환
- DAYNAME, DAYOFMONTH, DAYOFWEEK, WEEK
    - 날짜의 요일을 영어로 반환
    - 날짜의 일 반환
    - 날짜의 요일을 숫자로 반환
    - 날짜가 해당 연도에 몇 번째 주인지 반환
- HOUR, MINUTE, SECOND
    - 시간의 시, 분, 초 반환
- DATE_FORMAT
    - 날짜 / 시간의 형식을 형식으로 바꿔 반환

- DATEDIFF, TIMEDIFF
    - 두 날짜의 차이 반환
    - 두 시간의 차이 반환
