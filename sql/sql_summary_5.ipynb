# 데이터 그룹화 하기

## GROUP BY
- 컬럼에서 동일한 값을 가지는 로우를 그룹화하는 키워드
- `GROUP BY [컬럼 이름]` 형식으로 사용
- 그룹 별 데이터를 집계할 때 사용
- GROUP BY가 쓰인 쿼리의 SELECT 절에는 **GROUP BY 대상 컬럼**과 **그룹 함수**만 사용 가능
    - **(만약, GROUP BY 대상 컬럼이 아닌 컬럼을 SELECT 하면, 에러가 발생)**
- 여러 컬럼으로 그룹화도 가능, 키워드 뒤에 [컬럼]을 N개 입력
- SELECT절의 컬럼 순서 번호로도 그룹화가 가능

## HAVING
- 가져올 데이터 그룹에 조건을 지정해주는 키워드
- `HAVING 조건식` 형식으로 사용
- 조건식이 True가 되는 그룹만 선택
- HAVING절의 조건식에서는 그룹 함수를 활용

## 그룹함수
- 집계할 컬럼 이름은 그룹의 기준이 되는 컬럼 이름과 같아도 되고, 같지 않아도 됨
- GROUP BY가 없는 쿼리에서도 사용 가능, 이때는 전체 로우에 함수가 적용

### COUNT
- 그룹의 값 수를 세는 함수
- `COUNT(컬럼)` 형식으로 SELECT, HAVING 절에서 사용
- `COUNT(1)`은 하나의 값을 1로 세어주는 표현으로 그룹의 개수를 세어줌

### SUM
- 그룹의 합을 계산하는 함수
- `SUM(컬럼)` 형식으로 SELECT, HAVING 절에서 사용

### AVG
- 그룹의 평균을 계산하는 함수
- `AVG(컬럼)` 형식으로 SELECT, HAVING 절에서 사용

### MIN
- 그룹의 최솟값을 반환하는 함수
- `MIN(컬럼)` 형식으로 SELECT, HAVING 절에서 사용

### MAX
- 그룹의 최대값을 반환하는 함수
- `MAX(컬럼)` 형식으로 SELECT, HAVING 절에서 사용
