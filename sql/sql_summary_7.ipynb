# 테이블 합치기

## JOIN
- 테이블을 합칠 때 사용하는 키워드
- 같은 의미를 가지는 컬럼의 값을 기준으로 테이블을 합칠 때 사용하는 키워드



### 
1. INNER JOIN
- 기본 JOIN으로, DEFAULT 값
- 두 테이블 모두에 있는 값만 합치기

2. LEFT, RIGHT JOIN
- 왼쪽 & 오른쪽에 있는 값만 합치기

3. OUTER JOIN - **MySQL 키워드에 없음**
- 두 테이블에 있는 모든 값 합치기
- LEFT & RIGHT를 활용하여 `UNION` 키워드 사용 (UNION 키워드는 두 쿼리의 결과를 중복 제거하고 합쳐서 보여주는 집합 연산자)

4. CROSS JOIN
- 두 테이블에 있는 모든 값을 각각 합치기
- `ON` 키워드가 없어도 됨

5. SELF JOIN **(반드시 AS를 사용하여 별명을 만들어야 ERROR가 발생하지 않음)
- 같은 테이블에 있는 값 합치기
