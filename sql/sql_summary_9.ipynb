# 조건에 조건 더하기

- `하나의 쿼리 내 포함된 또 하나의 쿼리`를 의미
- 서브 쿼리는 반드시 `괄호 안에` 있어야 함
- `SELECT, FROM, WHERE, HAVING, ORDER BY` 절에서만 사용 가능 `INSERT, UPDATE, DELETE` 문에서도 사용 가능
- 서비쿼리에는 `;`를 붙이지 않아도 됨

## SELECT
- `스칼라 서브쿼리` 라고도 함
- SELECT절의 서브 쿼리는 반드시 결과값이 `하나의 값`

## FROM 
- `인라인 뷰 서브쿼리` 라고도 함
- FROM절의 서브 쿼리는 반드시 결과값이 `하나의 테이블`
- 서브 쿼리로 만든 테이블은 반드시 **별명**을 가져야 함

## WHERE
- `중첩 서브쿼리` 라고도 함
- WHERE절의 서브 쿼리는 반드시 결과값이 `하나의 컬럼` **(EXISTS 제외)**
- 연산자와 함께 사용 **(비교 연산자만 사용시, WHERE절의 서브 쿼리는 반드시 결과값이** `하나의 값`**이어야 함)**
